```python
import requests
from bs4 import BeautifulSoup
import time
import random
import pandas as pd
import re
from unidecode import unidecode
import pickle
note=[]
texte=[]
head={
"Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
"User-Agent":"Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:88.0) Gecko/20100101 Firefox/88.0"}
for i in range(1,800):
    url='https://www.amazon.fr/quatre-accords-tolt%C3%A8ques-libert%C3%A9-personnelle/product-reviews/2889116549/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber={}'.format(str(i))
    r=requests.get(url,headers=head)
    print(url,r.status_code)
    soup=BeautifulSoup(r.content)
    
    for elem in soup.findAll('i',attrs={"data-hook":"review-star-rating"}):
        note.append(elem.text.split(',')[0])

    for elem in soup.findAll('span',attrs={"data-hook":"review-body"}):
        texte.append(elem.text.strip())

    print("iteration : ",i,"/ longueur liste ",len(note),len(texte))
    
    if (len(note),len(texte)) != (i*10,i*10):
        m=min(len(note),len(texte))
        note=note[:m]
        texte=texte[:m]
        break

```

# Nettoyage

```python
df=pd.read_csv('Tolteques.csv')
```

```python
print(df.shape)
df.drop_duplicates(subset=['Texte'],inplace=True)
df.shape
```

```python
seuil_mot=10
def nb_mot(texte):
    return len(texte.split(' '))

df['Longueur']=df['Texte'].apply(nb_mot)
df=df[df['Longueur']>seuil_mot]
```

```python
df['Note'].value_counts()

# Binarisation 0 (1,2,3) / 1 (4,5)
df.loc[df[df['Note'].astype(int)<3].index,'Note']=0
df.loc[df[df['Note'].astype(int)>3].index,'Note']=1
df=df[df['Note']!=3]
# Je prends autant de texte positif que negatif (don j'identifie le label le moins present)
m=df['Note'].value_counts().min()
Comments=pd.concat([df[df['Note'].astype(int)==1].sample(m),df[df['Note'].astype(int)==0].sample(m)])
```

```python
Comments['Note'].value_counts()
```

```python
def nettoyage(texte):
    lem=pickle.load(open('lemmatize_fr.pickle','rb'))
    #Je charge la liste de stop words
    with open('stop-w_fr.txt') as f:
        s_w=f.read().split('\n')

    tmp=[]
    # Normalisation : accents et minuscule
    texte=unidecode(texte.lower())
    p='[a-z]{2,}'
    for mot in re.findall(p,texte):
        if mot in s_w:
            continue
        else:
            tmp.append(mot)
    return ' '.join(tmp)

nettoyage("élément à modifier chez Mr Bean")
```

```python
negatif=[nettoyage(elem) for elem in Comments[Comments['Note']==0]['Texte']]
positif=[nettoyage(elem) for elem in Comments[Comments['Note']==1]['Texte']]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#On prend les mots par groupe de 2,3 et 4
vectorizer = TfidfVectorizer(ngram_range=(2, 4))
#On utilise le vectorizer sur les commentaires 
X1= vectorizer.fit_transform(df['Texte'])
tmp=pd.DataFrame(X1.toarray(),columns=vectorizer.get_feature_names())
```

```python
X=pd.DataFrame(tmp)
X=X[1:1000]
y=df['Note']
y=y[1:1000]
```

```python
y
```

```python
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
import sklearn

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

clf = svm.LinearSVC()
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)


from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, X_train, y_train, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))


print("Train score : ",clf.score(X_train, y_train))
print("Test score : ",clf.score(X_test, y_test))
y_pred=clf.predict(X_test)
confusion_matrix(y_test, y_pred)
print("precision : ",sklearn.metrics.precision_score(y_test, y_pred))
print("F1 score : ",sklearn.metrics.f1_score(y_test, y_pred))
```
